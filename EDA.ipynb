```
student %>% map_df(~ sum(is.na(.)))
```

```
student %>% count(學校類型)
```

```
#台灣大學類型的圖 x=類型,y=數量
student %>%
  count(學校類型) %>%
  ggplot(aes(x = 學校類型, y = n, fill = 學校類型)) +
  geom_col() +
  geom_text(aes(label = n), vjust = -0.25) +
  labs(title = "台灣大學的類型", x = "類型", y = "數量") + theme(legend.position = "none")
```

```
#台灣大學的在籍外國生 x=數量,y=正式修讀學位外國生
student %>%
  count(學位生_正式修讀學位外國生) %>%
  ggplot(aes(x = n, y = 學位生_正式修讀學位外國生, fill = n)) +
  geom_col() +
  geom_text(aes(label = n), vjust = -0.25) +
  labs(title = "台灣大學的在籍外國生", x = "數量", y = "正式修讀學位外國生") + theme(legend.position = "none")
```
