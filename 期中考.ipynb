```
#用點狀圖表示助攻和比賽場數
Assists<-data.frame(Assists = factor(rep(letters[1:3],each=10)),gamesplayed=rnorm(30))
ggplot(Assists,aes(x=Assists,y=gamesplayed)) +geom_point()
```

```
#用點狀散佈圖表示助攻和比賽場數
#ggplot做不出來
NBA1718<-fetch_NBAPlayerStatistics("17-18")
qplot(Assists,GamesPlayed,data=NBA1718)
```

```
#用散佈圖表示助攻和比賽場數
qplot(Assists,GamesPlayed,data=NBA1718,geom=c("point","smooth"))
```

```
#用長條圖表示比賽場數
qplot(GamesPlayed,data=NBA1718,fill=Position)
```
